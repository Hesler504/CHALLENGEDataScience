### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Ingreso por Tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total tienda 1: {ingreso_tienda1}")
print(f"Ingreso total tienda 2: {ingreso_tienda2}")
print(f"Ingreso total tienda 3: {ingreso_tienda3}")
print(f"Ingreso total tienda 4: {ingreso_tienda4}")
```

# 2. Ventas por categoría

```python
# Ventas por categoría en cada tienda
ventas_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)

print("Ventas por categoría - Tienda 1:")
print(ventas_categoria_tienda1)
print("\nVentas por categoría - Tienda 2:")
print(ventas_categoria_tienda2)
print("\nVentas por categoría - Tienda 3:")
print(ventas_categoria_tienda3)
print("\nVentas por categoría - Tienda 4:")
ventas_categoria_tienda4
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio de clientes por tienda
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio tienda 1: {calificacion_promedio_tienda1}")
print(f"Calificación promedio tienda 2: {calificacion_promedio_tienda2}")
print(f"Calificación promedio tienda 3: {calificacion_promedio_tienda3}")
print(f"Calificación promedio tienda 4: {calificacion_promedio_tienda4}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

def analyze_products(df, store_name):
  # Productos más vendidos
  top_products = df.groupby('Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False).head(5)
  print(f"\nTop 5 productos vendidos en {store_name}:")
  print(top_products)

  # Productos menos vendidos
  bottom_products = df.groupby('Producto')['Cantidad de cuotas'].sum().sort_values(ascending=True).head(5)
  print(f"\nBottom 5 productos vendidos en {store_name}:")
  print(bottom_products)

  # Visualización
  plt.figure(figsize=(10, 5))
  plt.bar(top_products.index, top_products.values)
  plt.xlabel("Producto")
  plt.ylabel("Cantidad Vendida")
  plt.title(f"Top 5 productos vendidos en {store_name}")
  plt.xticks(rotation=45, ha="right")
  plt.tight_layout()
  plt.show()

  plt.figure(figsize=(10, 5))
  plt.bar(bottom_products.index, bottom_products.values)
  plt.xlabel("Producto")
  plt.ylabel("Cantidad Vendida")
  plt.title(f"Bottom 5 productos vendidos en {store_name}")
  plt.xticks(rotation=45, ha="right")
  plt.tight_layout()
  plt.show()

# Análisis para cada tienda
analyze_products(tienda, "Tienda 1")
analyze_products(tienda2, "Tienda 2")
analyze_products(tienda3, "Tienda 3")
analyze_products(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Costo de envío promedio por tienda
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio tienda 1: {costo_envio_promedio_tienda1}")
print(f"Costo de envío promedio tienda 2: {costo_envio_promedio_tienda2}")
print(f"Costo de envío promedio tienda 3: {costo_envio_promedio_tienda3}")
print(f"Costo de envío promedio tienda 4: {costo_envio_promedio_tienda4}")
```
